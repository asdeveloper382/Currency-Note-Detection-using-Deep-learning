```python
import torch
import cv2
import numpy as np
from PIL import Image
from IPython.display import display
from yolov5.models.common import DetectMultiBackend
from yolov5.utils.general import non_max_suppression, scale_boxes
from yolov5.utils.torch_utils import select_device
from collections import defaultdict

# Load the model
model_path = "/content/drive/MyDrive/best.pt"
device = select_device("cuda" if torch.cuda.is_available() else "cpu")
model = DetectMultiBackend(model_path, device=device)
model.model.float()
model.eval()

# Load and preprocess image
image_path = "/content/drive/MyDrive/prre/50224062_344879809440340_7212890943584731136_n_jpg.rf.89e872c303cc4d32b25473c4ed14021c.jpg"
img = cv2.imread(image_path)
assert img is not None, "Image not found"
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Resize and normalize
img_resized = cv2.resize(img_rgb, (640, 640))
img_tensor = torch.from_numpy(img_resized).permute(2, 0, 1).float() / 255.0
img_tensor = img_tensor.unsqueeze(0).to(device)

# Run inference
with torch.no_grad():
    pred = model.model(img_tensor)

# Apply NMS
pred = non_max_suppression(pred, conf_thres=0.5, iou_thres=0.4)[0]

# Detection summary
detections_per_class = defaultdict(int)
confidences = []

# Rescale boxes to original image size
scale_factor = img.shape[1] / 640  # Assuming width and height are equal in resizing
for det in pred:
    x1, y1, x2, y2, conf, cls = det.cpu().numpy()
    x1, y1, x2, y2 = [int(x * scale_factor) for x in [x1, y1, x2, y2]]
    label = f"{model.names[int(cls)]}: {conf:.2f}"

    # Draw bounding boxes and label
    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
    cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Count and confidence
    detections_per_class[model.names[int(cls)]] += 1
    confidences.append(conf)

# Print detection info
print("\nğŸ” Detection Summary:")
total_detections = sum(detections_per_class.values())
for cls, count in detections_per_class.items():
    print(f"- {cls}: {count} detections")
print(f"ğŸ“¦ Total Detections: {total_detections}")
if confidences:
    avg_conf = sum(confidences) / len(confidences)
    print(f"âœ… Average Confidence (proxy for accuracy): {avg_conf:.2f}")
else:
    print("âš ï¸ No detections found.")

# Display image
img_display = Image.fromarray(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
display(img_display)
```
